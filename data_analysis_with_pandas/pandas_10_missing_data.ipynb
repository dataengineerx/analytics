```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hpi_data_df = pd.read_pickle('data/fmac_all_states.pickle')
```

```python
hpi_data_df['TX1yr'] = hpi_data_df['TX'].resample('A').mean()
hpi_data_df[['TX','TX1yr']]
hpi_data_df.fillna(method="ffill", inplace=True)
hpi_data_df[['TX','TX1yr']]
hpi_data_df[['TX','TX1yr']].plot()
```

```python
hpi_data_df['TX1yr'] = hpi_data_df['TX'].resample('A').mean()
hpi_data_df[['TX','TX1yr']]
hpi_data_df.fillna(method="bfill", inplace=True)
hpi_data_df[['TX','TX1yr']]
hpi_data_df[['TX','TX1yr']].plot()
```

```python
hpi_data_df['TX1yr'] = hpi_data_df['TX'].resample('A').mean()
hpi_data_df[['TX','TX1yr']]
hpi_data_df.fillna(value=-9999, inplace=True,limit=len(hpi_data_df)//4,axis=1)
hpi_data_df[['TX','TX1yr']]
hpi_data_df[['TX','TX1yr']].plot()
```
