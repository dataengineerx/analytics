```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hpi_data_df = pd.read_pickle('data/fmac_all_states.pickle')
hpi_data_df.plot()
plt.legend().remove()
```

```python
for state in pd.read_pickle('data/states.pickle'):
    hpi_data_df[state] = (hpi_data_df[state] - hpi_data_df[state][0]) / hpi_data_df[state][0] * 100.0
hpi_data_df.plot()
plt.legend().remove()
```

```python
import quandl 
import os
def hpi_benchmark():
    benchmark = quandl.get("FMAC/HPI_USA", authtoken=os.environ.get('QUANDL_API_KEY'))
    benchmark["United States"] = (benchmark["NSA Value"] - benchmark["NSA Value"][0]) / benchmark["NSA Value"][0] * 100.0
    return benchmark[["United States"]]

```

```python
figure = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))
benchmark = hpi_benchmark()
hpi_data_df.plot(ax=ax1)
benchmark.plot(ax=ax1, color='k', linewidth=10)
plt.legend().remove()
```
