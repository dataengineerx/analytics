```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hpi_data_df = pd.read_pickle('data/fmac_all_states.pickle')
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))
TX1ye = hpi_data_df['TX'].resample('A').ohlc()
print(TX1ye.head())
hpi_data_df['TX'].plot(label='Monthly TX HPI')
TX1ye.plot(label='Yearly TX HPI')
plt.legend(loc=4)
plt.show()
```

```python
hpi_data_df['TX1yr'] = hpi_data_df['TX'].resample('A').mean()
hpi_data_df[['TX','TX1yr']]
hpi_data_df.dropna(how="all",  inplace=True)
hpi_data_df[['TX','TX1yr']].plot()
```
