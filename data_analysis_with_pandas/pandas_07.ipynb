```python
import os
import quandl
import pandas as pd
API_KEY = os.environ.get('QUANDL_API_KEY')
fmac_df = quandl.get('FMAC/HPI', authtoken=API_KEY)
fmac_df.to_pickle('fmac_all_states.pickle')
```

```python
fiddle_states = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states')[0]
states_df = (fiddle_states['Flag, name and postal abbreviation [1]']
                    .rename(columns={'Flag, name and postal abbreviation [1]': 'State', 'Flag, name and postal abbreviation [1].1': 'Code'})
                    )
states_df["Code"].to_pickle('states.pickle')
```
